```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Feigenbaum(x,r):
    return(x*(1-x)*r)
def data(x0,R):
    XN=np.empty((1000,np.size(R)))
    j=0
    for r in R:
        #print(r)
        x=x0
        for i in range(0,1000):
            #print(x)
            xn=Feigenbaum(x,r)
            XN[i,j]=xn
            x=xn
        j=j+1
    return(XN)
```

```python
R0=np.arange(0,4,0.0001)
Y=data(0.2,R0) 
print(Y)
F=np.transpose(Y)
%matplotlib notebook
#fig = plt.figure(figsize=(6,5))
plt.plot(R0,F[:,-100:],'.',markersize=0.01)
plt.show()
plt.savefig("Feigenbaum.pdf")
```
